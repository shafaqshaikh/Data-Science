```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
x = np.linspace(0,5,11)
```

```python
y = x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x , y , 'r')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('title')
```

```python
plt.subplot(1,2,1)
plt.plot(x , y , 'r')
plt.subplot(1,2,2)
plt.plot(y , x , 'b')
```

```python
#object oriented
fig = plt.figure()
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1 , 0.1 , 0.8, 0.8])
axes2 = fig.add_axes([0.1 , 0.3 , 0.2, 0.2])
axes1.plot( x , y)
axes2.plot( x , y)
axes1.set_title('title')
axes2.set_title('title')
```

```python
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.8,0.8])
ax.plot(x , y , color = 'black' , linewidth=2, linestyle='steps' , marker='o',markersize=20,markerfacecolor='red',markeredgewidth=3,markeredgecolor='blue')
```
